```python
y = np.linspace(0,1,50)
s_array = np.array([2,4,16])
t_array = np.linspace(0,1,5)
def conc_schlick(x,slope,threshold): # Barron 2020
    epsilon = np.power(0.1,10)
    s = slope
    t = threshold
    Ps = ((1.0 - s) + s*np.log(s))/((1.0-s)**2)
    PhiR = (2*t-1.0)*Ps + (1-t) 
    h = 1-x
    x_low = h[h<t]
    x_high = h[h>=t]
    ans_low = t*x_low /( x_low + s * (t - x_low) + epsilon )
    ans_high = (1 - t) * (x_high - 1) / (1-x_high-s*(t-x_high) + epsilon ) + 1
    Cz = np.concatenate((ans_high,ans_low), axis = None )
    return Cz/PhiR

fig,axes = ff.subplot2gridlist(len(s_array),len(t_array),(A4['width'], A4['half']))
ff.adjustTicks(axes,[-0.5,7],[-0.1,1.1],xticks=[0,2,4,6],xlabels=[0,2,4,6],yticks=[0,0.5,1],ylabels=[0,0.5,1])
for i,s in enumerate(s_array):
    for j,t in enumerate(t_array):
        x = conc_schlick(y,s_array[i],t_array[j])
        axes[j][i].plot(x,y,zorder=1)
        axes[j][i].scatter(conc_schlick((1-t_array[j]),s_array[i],t_array[j]),(1-t_array[j]),s=2,color='red',zorder=2)
        axes[j][i].grid()
#         axes[j][i].axhline(0,color='gray',alpha=1)
#         axes[j][i].axvline(0,color='gray',alpha=1)
        if i == 0:
#             axes[j][i].set_title(r'$t=$'+str(t))
            axes[j][i].text(0.5, 1.1, r'$t=$'+str(t),
                    horizontalalignment='center',
                    verticalalignment='center',
                    rotation='horizontal',
                    transform=axes[j][i].transAxes)
        if j == 4:
            axes[j][i].text(1.1, 0.5, r'$s=$'+str(s),
                    horizontalalignment='center',
                    verticalalignment='center',
                    rotation='vertical',
                    transform=axes[j][i].transAxes)
fig.text(0.5, 0.00, r'$ \langle \phi \rangle / \Phi $', ha='center')
fig.text(0.00, 0.5, r'$z/h$', va='center', rotation='vertical')
plt.tight_layout()
plt.show()
plt.close()
```
